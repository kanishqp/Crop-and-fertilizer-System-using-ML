```python
import seaborn as sns
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(crop[i], ax=subplot, kde=True)
    subplot.set_title(i)
plt.tight_layout()
plt.show() 
```

Bivariate Analysis

```python
#plot a scatter graph
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3,3, figsize=(15,15))
for i,subplot in zip(features,ax.flatten()):
    sns.scatterplot(x=1,y='label',data=crop,ax=subplot)
    subplot.set_title(i)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_scatter_matrix(data, features):
    fig, axes = plt.subplots(len(features), 1, figsize=(15, 15))
    for i, feature in enumerate(features):
        sns.scatterplot(x=1, y='label', data=data, ax=axes[i])
        axes[i].set_title(features)
    plt.tight_layout()
    plt.show()

plot_scatter_matrix(crop, features= ['P', 'K', 'temperature', 'humidity', 'ph', 'rainfall', 'label'])
```

```python
#finding outliers
fig , ax = plt.subplots(3,3,figsize=(15,15))
for i ,subplot in zip(features ,ax.flatten()):
    sns.boxplot(crop[i],ax=subplot )
plt.tight_layout()
plt.show()
```
